```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки
# зі спеціалістами, які вказали мову програмування Python

python_data = pd.read_csv("/content/sample_data/2017_jun_final.csv", sep=',' , encoding='utf-8')
```

```python
python_data = python_data[python_data['Язык.программирования'] == 'Python']
python_data.shape
```

```python
job_avg = python_data.groupby('Должность').agg({'Зарплата.в.месяц': ['min', 'max']})
job_avg
```

```python
def fill_avg_salary(row):
  min = row['Зарплата.в.месяц']['min']
  max = row['Зарплата.в.месяц']['max']
  print(row)
  return (max + min) / 2
```

```python
job_avg['avg'] = job_avg.apply(fill_avg_salary, axis=1)
job_avg
```

```python
job_avg['avg'].describe()
```

```python
job_avg.to_csv("/content/2017_jun_final_DZ.csv", index=False)
```

```python
#csv_f[csv_f['Пол'] == 'женский'].loc[:, ['Должность','Пол']]
gender_counts_w = csv_f[csv_f['Пол'] == 'женский'].groupby('Должность')['Пол'].value_counts()
gender_counts_m = csv_f[csv_f['Пол'] == 'мужской'].groupby('Должность')['Пол'].value_counts()
merged_gender_counts = pd.merge(gender_counts_w, gender_counts_m, on='Должность', how='outer').fillna(0)
```

```python
merged_gender_counts
```

```python
from matplotlib import pyplot as plt
merged_gender_counts['Пол_y'].plot(kind='line', figsize=(8, 4), title='Пол_y')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title('Кількість чоловіків по должностям')
plt.xticks(rotation=45)
```

```python
from matplotlib import pyplot as plt
merged_gender_counts['Пол_x'].plot(kind='line', figsize=(8, 4), title='Пол_x')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title('Кількість жінок по должностям')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(15, 8))

plt.plot(merged_gender_counts.index, merged_gender_counts['Пол_x'], label='Women', marker='o')
plt.plot(merged_gender_counts.index, merged_gender_counts['Пол_y'], label='Men', marker='o')
plt.xlabel('Должность')
plt.ylabel('Кількість осіб')
plt.title('Розподіл кількості осіб за посадами та статтю')
plt.xticks(rotation=90)
plt.legend()

plt.tight_layout()
plt.show()
```
