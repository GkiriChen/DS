```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8",
                  match="Коефіцієнт народжуваності в регіонах ", decimal=',', thousands=' ')[0]

numeric_columns = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']

for column in numeric_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')
print(df)
```

```python
#tmp[0].head()
print(df.dtypes)
```

```python
df_1 = df.copy()
print(df.shape)
```

```python
print(df_1)
print('+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-')
print()
df_1 = df_1.drop(df_1.index[27])
print(df_1)
```

```python
df_1.fillna(df_1.mean(), inplace=True)
```

```python
average = df_1.mean()
print(f"середне народжуваність у 2019 році = {average['2019']}")
df_sort = df_1[df_1['2019'] > average['2019']]
df_sort.loc[:, ['Регіон', '2019']]
```

```python
import seaborn as sns

#data = sns.load_dataset("Регіон")
sns.set_style("whitegrid")
plt.figure(figsize=(12, 10))
sns.lineplot(data=df_1.melt(id_vars=['Регіон'], var_name='Рік', value_name='Коефіцієнт народжуваності'), x='Рік', y='Коефіцієнт народжуваності', hue='Регіон')

plt.title('Народжуваність по регіонах у 1950 - 2019')
plt.xlabel('Рік')
plt.ylabel('Коефіцієнт народжуваності')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data = df_1.loc[:, ['Регіон', '1950']]

plt.figure(figsize=(10, 6))
sns.lineplot(x=data['Регіон'], y=data['1950'])
plt.title(f'Народжуваність по регіонах у 1950 році')
plt.xlabel('Регіон')
plt.ylabel('Коефіцієнт народжуваності')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
